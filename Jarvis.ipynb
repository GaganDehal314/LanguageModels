# Streamlit App for an AI personal assistant

I've created a streamlit app that connects to a MySQL Database and runs queries on the database tables. The app leverages a Google PaLM model to run the queries and wrap the SQL result in any suitable generated text to seem more like a personal assistant.

## Chat page

You can see the sample questions that I've asked the model. The layout of the answer changes as per the specific instructions that we provide.

The model can also use its knowledge of natural language to rearrange data without making any changes to the SQL query it generates. Here, I asked the model to sort the data and put people with coding-based jobs on top. It did its best to do so. However, the sample data didn't have too many connections that did have the desired roles.

### Use cases

On a day-to-day basis, I can use my AI assistant to contact my connections on LinkedIn. Instead of aimlessly going around the website and finding connections that have the roles I'd like to target; I can simply sort them in my app and make a list of people I'd like to aim at reaching out. 

The same can be done for employers as well. Finding people who are into Talent Acquisition or are in an executive role in a company would be a good fit for my goal.
